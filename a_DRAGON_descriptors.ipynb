```python
import pandas as pd ; pd.set_option('display.max_columns', 500)
import numpy as np
from numpy import random ; random.RandomState(seed = 69)

import matplotlib.pyplot as plt
%matplotlib inline
import ipywidgets as widgets
```

# Take DRAGON output and convert it to a dataframe

```python
const_df = pd.read_csv('filtered_organics_kh.csv')
desc_df = pd.read_table('dragon_output_filtered_organics.txt')
desc_df.drop('No.', axis=1, inplace=True)
```

```python
index_list = []
for i in range(0, len(desc_df)):
    index_list.append(i)
desc_df['Index'] = index_list
```

Get index for `desc_df` to re-order species after turning them into sets and back again

```python
const_df_smiles = list(const_df.SMILES)
desc_df_smiles = list(desc_df.NAME)
print(len(desc_df_smiles), len(const_df_smiles))
print(len(desc_df_smiles) - len(const_df_smiles))
```

```python
const_df.set_index('SMILES', inplace=True)
desc_df.set_index('NAME', inplace=True)
```

# Find species which DRAGON could not calculate for

```python
const_set = set(const_df_smiles)
desc_set = set(desc_df_smiles)
missing_list = (const_set - desc_set)
to_use = list(const_set - set(missing_list))
const_df_2 = const_df.loc[to_use]
desc_df_2 = desc_df.loc[to_use]
df = pd.concat([const_df_2, desc_df_2], axis=1)
```

```python
df_sorted = df.sort_values(by=['Index'])
df_sorted.head()
```

```python
df_sorted.to_csv('filtered_organics_desc_kh.csv')
```

# Principal Component Analysis

```python
from sklearn.preprocessing import StandardScaler
pca_init = pd.concat([desc_df_2, const_df_2.Kh_mean], axis=1)
pca_init.dropna(axis=0, how='any', inplace=True)
x = pca_init.drop('Kh_mean', axis=1).values
y = pca_init.Kh_mean.values #target

#standardizing features
#converts to unit scale (mean=0, variance=1)
# = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA

#no particular meaning assigned to each component
#just reducing dimensionality
pca = PCA(n_components=50)

principal_components = pca.fit_transform(x)

principal_df = pd.DataFrame(data=principal_components)
principal_df.shape
```

```python
principal_df['Constant'] = list(y)
```

```python
principal_df.to_csv('pca_2_12.csv')
```
